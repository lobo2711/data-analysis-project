```python
#Importar librerias necesarias

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Importamos base de datosb
df =pd.read_csv("/content/10k_anime_data.csv")
pd.set_option('display.max_columns', None)
df.head(5)
```

```python
# modificamos la base de datos removiendo columnas que no necesitamos
df = df[['title', 'title_english', 'aired_from',
       'aired_to', 'status', 'episodes', 'rating', 'rank',
       'popularity', 'members', 'favorites', 'score', 'scored_by', 'genres',
       'themes', 'demographics']]

df.head(5)
```

```python
# Columna demographics tiene valores Nan, podriamos eliminarlos con un dropna pero eso nos haria perder datos, en su lugar vamos a remplazarlos con 'missing'
df['demographics'] = df['demographics'].fillna('missing')
df.head(5)
```

```python
#Reisamos si aun quedan datos nulos NaN
df.isnull().sum()
```

```python
#Se encuentran datos nulos en la columna de fecha 'aired_from' esto nos puede generar problemas luego si queremos usarla la columna para algun grafico asi que procedemos a retirarlos
df = df.dropna(subset=['aired_from'])
df.isnull().sum()
```

```python
#Ahora con los datos limpios en 'aired_from' procedemos a modificar la columna a fecha y a crear una columna nueva con solo el año de salida con nombre 'year'
df['aired_from'] = pd.to_datetime(df['aired_from'])
df['year'] = df['aired_from'].dt.year
df.head(5)
```

```python
# Ahora con los datos necesarios podemos empezar con graficos para el analisis de datos empezamos con un grafico para verificar la tendencia entre la popularidad y el score
# Aca podemos interpretar que no necesariamente los animes mas populares gozan de un score alto, por el contrario muchas veces es al contrario
sns.scatterplot(x='popularity', y='score', data=df)
```

```python
# Seguido queria hacer un histograma de la cantidad de animes creados atraves de los años
# podemos confirmar que atraves de los años los animes han ido ganando popularidad entre las personas con un incremento exponencial año con año y explotando apartir de la decada de 2000
sns.histplot(df['year'], bins=20)
```

```python
# Para el siguiente grafico queria verficar el tipo 'rating' de anime que ha predominado por decada
# Create a 'decade' column
df['decade'] = (df['year'] // 10) * 10

# Group by decade and rating, and count the number of titles
rating_decade_counts = df.groupby(['decade', 'rating']).size().reset_index(name='count')

# Create a grouped bar plot
plt.figure(figsize=(12, 6))
sns.barplot(data=rating_decade_counts, x='decade', y='count', hue='rating')
plt.xlabel('Decade')
plt.ylabel('Number of Titles')
plt.title('Number of Titles per Rating per Decade')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()

#Podemos verificar que el rating de PG-13 siempre ha predominado, apesar que todos los rating incrementaron bastante en el año 200o PG-13 predomino aun mas
```

```python
# Boxplot de la relacion entre rating con el score de los animes
sns.boxplot(x='rating', y='score', data=df)
plt.xticks(rotation=60)
plt.show()

# Podemos determinar que apesar de que PG-13 es el rating que predomina en los ultimos años la saturacion de productos han tambien producido una caida en la calidad siendo este rating el que tiene mas productos
#-muy por debajo de la media y siendo R-17 + el que mejores valoraciones obtiene
```

```python
# Heatmap para verificar correlaciones entre varios valores numericos, podemos verificar que popularidad y rank se relacionan mejor asi como miembros y favoritos
df_num =  df[['rank', 'popularity', 'members', 'favorites']]
plt.figure(figsize=(10,5))
corr= df_num.corr()
sns.heatmap(corr,cmap="BrBG",annot=True)
corr
```
